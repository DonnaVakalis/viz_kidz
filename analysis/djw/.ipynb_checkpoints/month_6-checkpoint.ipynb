```r
library(tidyverse)
library(data.table)

library(lubridate)

library(ggplot2)
library(ggthemes)
library(ggforce)
```

```r
options(repr.plot.width=6.5, repr.plot.height=5)
```

```r
aw = read.csv("../../dat/Amos_20190613_20200111.csv")
rw = read.csv("../../dat/Rilka_20190613_20200111.csv")
```

```r
aw$kid = "Amos"
rw$kid = "Rilka"
```

```r
kidz = rbind(aw, rw)
```

```r
head(kidz,3)
```

```r
str(kidz)
```

```r
# convert from factor to string
kidz$Date.and.Time = as.character(kidz$Date.and.Time)

# convert kid to factor
kidz$kid = as.factor(kidz$kid)
```

```r
# convert to datetime object with lubridate
kidz$Date.and.Time = lubridate::mdy_hm(kidz$Date.and.Time)
kidz$End.Time = lubridate::mdy_hm(kidz$End.Time)
```

```r
head(kidz,3)
```

```r
str(kidz)
```

```r
kidz %>% group_by(month=lubridate::floor_date(Date.and.Time, "month"), kid) %>%
   summarize(amount=sum(Quantity, na.rm = T)) %>%
    head()
```

```r
# create variable for first entry
birth = min(kidz$Date.and.Time)
```

```r
# add month label
kidz$month = NA

for (i in seq(1,6,1)){
    interval = lubridate::interval(birth+months(i-1), birth+months(i)) 
    
    for (j in seq(length(kidz$Date.and.Time))){
        if (kidz$Date.and.Time[j] %within% interval){
            kidz$month[j]=i
        }
    }
}         
         
```

```r
tail(kidz,3)
```

```r
# add week label

kidz$week = NA

for (i in seq(1,27,1)){
    interval = lubridate::interval(birth+weeks(i-1), birth+weeks(i)) 
    
    for (j in seq(length(kidz$Date.and.Time))){
        if (kidz$Date.and.Time[j] %within% interval){
            kidz$week[j]=i
        }
    }
}   
```

```r
saveRDS(kidz, "../../dat/combined_data_djw.rds")
```

```r
tail(kidz,3)
```

```r
# weekly consumption

weekly_consumption = kidz %>%
    filter(Activity == "Bottle" & week<27) %>%
    group_by(week, kid) %>%
    summarise(total = sum(Quantity, na.rm=T))
```

```r
options(repr.plot.width=7, repr.plot.height=4)
```

```r
ggplot(weekly_consumption, aes(x = week, y = total)) + 
  geom_area(aes(color = kid, fill = kid), 
            alpha = 0.5, position = position_dodge(0.8)) +
    theme_minimal()
```

```r
# weekly sleep

weekly_sleep = kidz %>%
    filter(Activity == "Sleep" & week<27 ) %>%
    group_by(week, kid) %>%
    summarise(total = sum(Duration..min., na.rm=T)/(60*7))
```

```r
options(repr.plot.width=7, repr.plot.height=4)
```

```r
ggplot(weekly_sleep, aes(x = week, y = total)) + 
  geom_area(aes(color = kid, fill = kid), 
            alpha = 0.5, position = position_dodge(0.8)) +
    coord_cartesian(ylim = c(13, 17.5)) +
    theme_minimal()
```

# Pie Chart

```r
# monthly consumption

monthly_consumption = kidz %>%
    filter(Activity == "Bottle") %>%
    group_by(month, kid) %>%
    summarise(total = sum(Quantity, na.rm=T))
```

```r
options(repr.plot.width=6, repr.plot.height=5)
```

```r
p1 <- ggplot( monthly_consumption, aes( x = total/2, y = total, fill=kid, width=total ) ) + 
      geom_bar( stat = "identity", position="fill", width=1, colour="white") + 
    coord_polar("y", start=0) +
      facet_wrap( ~ month )
p1 + theme_void()
```

```r
# monthly sleep

monthly_sleep = kidz %>%
    filter(Activity == "Sleep") %>%
    group_by(month, kid) %>%
    summarise(total = sum(Duration..min., na.rm=T))
```

```r
options(repr.plot.width=6, repr.plot.height=5)
```

```r
p1 <- ggplot( monthly_sleep, aes( x = total/2, y = total, fill=kid, width=total ) ) + 
      geom_bar( stat = "identity", position="fill", width=1, colour="white") + 
    coord_polar("y", start=0) +
      facet_wrap( ~ month )
p1 + theme_void()
```
